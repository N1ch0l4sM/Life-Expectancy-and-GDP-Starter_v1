```python
import matplotlib as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Load the data
df = pd.read_csv('all_data.csv')

df
```

```python
plot = sns.boxenplot(x='Country',y= 'Life expectancy at birth (years)',data=df)
plot.set_xticklabels(plot.get_xticklabels(), rotation=45,horizontalalignment='right')
```

```python
chile = df[df['Country'] == 'Chile']
china = df[df['Country'] == 'China']
germany = df[df['Country'] == 'Germany']
mexico = df[df['Country'] == 'Mexico']
usa = df[df['Country'] == 'United States of America']
zimbabew = df[df['Country'] == 'Zimbabwe']

# get min, max, mean of life expectancy for each country
chile_min = chile['Life expectancy at birth (years)'].min()
chile_max = chile['Life expectancy at birth (years)'].max()
chile_mean = chile['Life expectancy at birth (years)'].mean()
china_min = china['Life expectancy at birth (years)'].min()
china_max = china['Life expectancy at birth (years)'].max()
china_mean = china['Life expectancy at birth (years)'].mean()
germany_min = germany['Life expectancy at birth (years)'].min()
germany_max = germany['Life expectancy at birth (years)'].max()
germany_mean = germany['Life expectancy at birth (years)'].mean()
mexico_min = mexico['Life expectancy at birth (years)'].min()
mexico_max = mexico['Life expectancy at birth (years)'].max()
mexico_mean = mexico['Life expectancy at birth (years)'].mean()
usa_min = usa['Life expectancy at birth (years)'].min()
usa_max = usa['Life expectancy at birth (years)'].max()
usa_mean = usa['Life expectancy at birth (years)'].mean()
zimbabew_min = zimbabew['Life expectancy at birth (years)'].min()
zimbabew_max = zimbabew['Life expectancy at birth (years)'].max()
zimbabew_mean = zimbabew['Life expectancy at birth (years)'].mean()

chile_status = [chile_min, chile_max, chile_mean]
china_status = [china_min, china_max, china_mean]
germany_status = [germany_min, germany_max, germany_mean]
mexico_status = [mexico_min, mexico_max, mexico_mean]
usa_status = [usa_min, usa_max, usa_mean]
zimbabew_status = [zimbabew_min, zimbabew_max, zimbabew_mean]

# Create a dataframe
df_status = pd.DataFrame({'Country': ['Chile', 'China', 'Germany', 'Mexico', 'United States of America', 'Zimbabwe'], 'Life expectancy at birth (years)': [chile_status, china_status, germany_status, mexico_status, usa_status, zimbabew_status]})
df_status
```

```python
# Plot the data life expectancy china trough time
sns.lineplot(x='Year',y= 'Life expectancy at birth (years)',data=chile)
sns.lineplot(x='Year',y= 'Life expectancy at birth (years)',data=china)
sns.lineplot(x='Year',y= 'Life expectancy at birth (years)',data=germany)
sns.lineplot(x='Year',y= 'Life expectancy at birth (years)',data=mexico)
sns.lineplot(x='Year',y= 'Life expectancy at birth (years)',data=usa)
sns.lineplot(x='Year',y= 'Life expectancy at birth (years)',data=zimbabew)


```
